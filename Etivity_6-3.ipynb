```python
# Define the loss function and the optimizer
from textblob.classifiers import NaiveBayesClassifier
import nltk
nltk.download('punkt')

def naiveBayesclassifier(trainingset, testset):
    classifier = NaiveBayesClassifier(trainingset)
    for i in testset :
      print("document ", i, "           predicted class : ", classifier.classify(i[0]))
    print("Accuracy: ", classifier.accuracy(testset))

trainingset = [('London is the Capital of GB', 'GB'),
                ('Oxford is a city in GB', 'GB'), ('Dublin is the capital of Ireland', 'IE'), ('Limerick is a city in Ireland', 'IE')]
testset = [('University of Limerick', 'IE'),
('University College Dublin', 'IE'), ('Imperial college London', 'GB'), ('University of Oxford', 'GB'),
('Ireland & GB', 'GB')]

naiveBayesclassifier(trainingset, testset)
```

```python
import pandas as pd
text_df = pd.read_csv('bbc-text.csv', header=None, delimiter = ',', names = ['category', 'text'])
text_df = text_df.drop(0)
display(text_df.head(100))

print("total number of documents :", len(text_df))
print ()
print("Document per category")
print(text_df['category'].value_counts())
print()


# splitting text_df 80/20
split_index = int(len(text_df) * 0.8)


train_df = text_df[:split_index]
test_df = text_df[split_index:]

print("train size",len(train_df))
print("test size",len(test_df))

print("Training set Samples")
print(train_df)

print("Test set Samples")
print(test_df)


# changing the format ofthe text
text_list = train_df['text'].tolist()
category_list = train_df['category'].tolist()
train_tuples = list(zip(text_list, category_list))
# print for validation
print(train_tuples)
# calling the classifier
classifier = NaiveBayesClassifier(train_tuples)

k = []
for i in test_df['text'] :
    k.append(classifier.classify(i))
    # print(test_df['text'], k)
# adding the classified data to the test_tuples
test_list = test_df['text'].tolist()
test_tuples = list(zip(test_list, k))

# validation
print(test_tuples[:10])
# accuracy Printed
print("Accuracy :", classifier.accuracy(test_tuples))


```
