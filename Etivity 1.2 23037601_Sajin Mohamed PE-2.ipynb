```python
!wget http://www.gutenberg.org/files/100/100-0.txt
filePath = "/content/100-0.txt"
ShakespeareFile = open(filePath, "r")
ShakespeareContent = ShakespeareFile.read()
ShakespeareContent = ShakespeareContent.splitlines() # The splitlines() method splits a string into a list. The splitting is done at line breaks.
print(ShakespeareContent[0:50])
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer


t  = Tokenizer()
# Defining 4 document lists
fit_text = ShakespeareContent

t.fit_on_texts(fit_text)
print("total number of tokens", sum(t.word_counts.values()))
print("total number of types", len(t.word_counts))

print(f"most frequent Type in corpus is '{max(t.word_counts)}' and its frequency is '{max(t.word_counts.values())}'")


```

```python
from nltk.stem import PorterStemmer
import nltk
nltk.download('wordnet')
from nltk.stem.wordnet import WordNetLemmatizer


ps = PorterStemmer()
lemmatizer = WordNetLemmatizer()
```

```python
words = list(t.word_counts.keys())
stemmedWords = set()
for word in words :
  stemmedWord = ps.stem(word)
  stemmedWords.add(stemmedWord)

print(stemmedWords)
print("total number of types after stemming is ", len(stemmedWords))
```

```python


lemma = set()
for word in words:
  lemma.add(lemmatizer.lemmatize(word))
print("length after lemmatization is ", len(lemma))
```

As we can see the count of types in corpus is 30491 and after lemmatization it reduced to 27188 and after stemming it reduced further to 21336

```python
assertTrue(len(lemma) < len(stemmedWords) < len(t.word_counts))
```
