```python
trainingSet = [('London is the Capital of GB', 'GB'), ('Oxford is a city in GB', 'GB'), ('Dublin is the capital of Ireland', 'IE'), ('Limerick is a city in Ireland', 'IE')]
testSet = ["University Of Limerick", "University College Dublin", "Imperial College London", "University of Oxford", "Ireland & GB"]

megaDocGB = []
megaDocIE = []
total_count_of_documents_in_train = 0
total_count_of_documents_in_GB = 0
total_count_of_documents_in_IE = 0
# get count for calculating the prior and form the megadocument
for i in trainingSet:
    total_count_of_documents_in_train += 1
    if i[1] == 'GB':
        megaDocGB.append(i[0].lower())
        total_count_of_documents_in_GB += 1

    else:
        megaDocIE.append(i[0].lower())
        total_count_of_documents_in_IE += 1

priorGB = total_count_of_documents_in_GB/total_count_of_documents_in_train
priorIE = total_count_of_documents_in_IE/total_count_of_documents_in_train

print("prior of GB", priorGB)
print("prior of IE", priorIE)
print()
print("megaDocGB", megaDocGB)
print("megaDocIE", megaDocIE)
print()

# used for find the bag of words provided the megadocument
def find_BOW(megaDoc):
    count = {}
    words = ' '.join(megaDoc).split()

    for i in words:
        if i not in count:
            count[i] = 1
        else:
            count[i] += 1
    return count

GB_BOW = find_BOW(megaDocGB)
IE_BOW = find_BOW(megaDocIE)
print("GB_BOW = ", GB_BOW)
print("IE_BOW = ", IE_BOW)
print(type(GB_BOW))


# used to find the vocabulary count
listGB = ' '.join(megaDocGB).split()
listIE = ' '.join(megaDocIE).split() + listGB


vocabulary = {}

for i in listIE:
    vocabulary[i] = listIE.count(i)

# displaying the vocabulary
print("V = ", vocabulary)
print(f"|V| = {len(vocabulary)}")
print()

count = {}
words = ' '.join(find_BOW(megaDocGB)).split()
for i in words:
        if i not in count:
            count[i] = 1
        else:
            count[i] += 1

print(count)

# used for get the count of words
def get_map_count_for_each_class(word, BOW):
    count = 0
    if (word in BOW) :
      count += 1

    return count



def naiveBayesClassifier(trainingSet, testSet) :

  for i in testSet :
    print(f"Test Document = ({i})")
    word_cond_prob_GB = 1
    word_cond_prob_IE = 1
    for word in i.split():
      lower_word = word.lower()

      print("word = ", lower_word)
      count_of_word_in_GB = get_map_count_for_each_class(lower_word, GB_BOW)
      count_of_word_in_IE = get_map_count_for_each_class(lower_word, IE_BOW)

      number_of_words_in_GB = sum(GB_BOW.values())
      number_of_words_in_IE = sum(IE_BOW.values())

      print("wordCondProbabiltyGB = ", (count_of_word_in_GB + 1)/ (number_of_words_in_GB + (len(vocabulary))))
      print("wordCondProbabiltyIE = ", (count_of_word_in_IE + 1)/ (number_of_words_in_IE + (len(vocabulary))))
      word_cond_prob_GB = word_cond_prob_GB * (count_of_word_in_GB + 1)/ (number_of_words_in_GB + (len(vocabulary)))
      word_cond_prob_IE = word_cond_prob_IE * (count_of_word_in_IE + 1)/ (number_of_words_in_IE + (len(vocabulary)))

      print()

    print("DocProbGB = ", round(priorGB * word_cond_prob_GB,5))
    print("DocProbIE = ", round(priorIE * word_cond_prob_IE,5))

    if(priorGB * word_cond_prob_GB > priorIE * word_cond_prob_IE) :
        print("Inferred Class : GB" )
    else :
        print("Inferred Class : IE" )
    print("-"*90)



naiveBayesClassifier(trainingSet, testSet)
```
