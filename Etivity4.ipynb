```python
# define a generic method to get and display the count of unigram and bigram
def find_count_value(dataFrame, words, identifier) :

  # filter out dataframe that has the unigram to then select the count
      wordExistDf = dataFrame[dataFrame['word'] == words]

      if not wordExistDf.empty:
        # Retrieve the 'count' value from the filtered DataFrame
        dataFrame_count_value = wordExistDf['count'].iloc[0]
      return dataFrame_count_value
```

```python
import jellyfish
import pandas as pd
unigrams_df = pd.read_csv('count_1w.txt', header=None, delimiter = '\t', names = ['word', 'count'])
display(unigrams_df.head(100))

filteredResult = ['word', 'frequency', 'probability']
df = []
print(unigrams_df.head(100))

def nonWordSpellingCorrection(nonword) :
  unigram_count = unigrams_df['count'].sum()
  for value in unigrams_df['word']:
    if isinstance(value, str):
        distance = jellyfish.damerau_levenshtein_distance(nonword, value)
        if (distance == 1) :
          frequency = find_count_value(unigrams_df, value, 'frequency')
          new_row = [value, frequency, frequency/unigram_count]
          df.append(new_row)


nonWordSpellingCorrection("acress")
print(pd.DataFrame(df, columns=filteredResult))

```

```python
def nonWordSpellingCorrection(nonWord):

   unigrams_df['dist'] = unigrams_df['word'].apply(lambda x : jellyfish.damerau_levenshtein_distance(str(x),nonWord))
   unigrams_df['P(word)'] = unigrams_df['count']/unigrams_df['count'].sum()
   candidates = unigrams_df[unigrams_df['dist']==1]
   return candidates
```
